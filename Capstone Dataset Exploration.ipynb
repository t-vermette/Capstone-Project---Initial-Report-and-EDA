#### Cleaning Median Income 2024 Dataset and naming copy as median_income2024_tidy.
##### Transposing, resetting Index, Changing column names to 'Metro' and 'Income2024'.

```python
median_income2024_tidy = median_income2024.copy()
```

```python
def clean_education_row(df, row_number, column_name):
    """
    Extracts one education level from the ACS education table
    and returns a tidy Metro/value DataFrame.
    """

    temp = df.iloc[[row_number]].T.reset_index()

    temp.columns = ['Metro', column_name]

    temp = temp.iloc[1:].copy()

    temp = temp[
        temp['Metro'].str.contains('Estimate')
    ].copy()

    temp['Metro'] = (
        temp['Metro']
        .str.replace(' Metro Area!!Estimate', '', regex=False)
    )

    temp[column_name] = (
        temp[column_name]
        .astype(str)
        .str.replace(',', '', regex=False)
    )

    temp[column_name] = pd.to_numeric(temp[column_name])

    return temp
```

## Data Dictionary

| Column | Description |
|---------|-------------|
| **RegionID** | Unique Zillow identifier for each metropolitan region. |
| **SizeRank** | Zillow ranking of metropolitan areas by population size. Lower values indicate larger metros. |
| **Metro** | Zillow metropolitan area name. |
| **RegionType** | Geographic region classification provided by Zillow (e.g., Metro). |
| **StateName** | State in which the metropolitan area is located. |
| **Date** | Observation date (December 2024 snapshot). |
| **HomeValue** | Zillow Home Value Index (ZHVI), representing the typical home value within the metropolitan area. |
| **Year** | Calendar year corresponding to the observation. |
| **CensusMetro** | Standardized Census metropolitan area name used to merge ACS datasets with Zillow data. |
| **Income2024** | Median household income (2024 American Community Survey). |
| **Population2024** | Total metropolitan population (2024 American Community Survey). |
| **CollegeEducated2024** | Number of residents age 25 and older with a bachelor's degree or higher (2024 American Community Survey). |
| **CollegeEducatedPct** | Proportion of residents with a bachelor's degree or higher. Calculated as `CollegeEducated2024 / Population2024` in the current analysis. |
